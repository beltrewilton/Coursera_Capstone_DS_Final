```python
import math

# inputs
radius = 25.0 # m - the following code is an approximation that stays reasonably accurate for distances < 100km
jump = 2 # jump to center 
centerLat = 39.9624589 # latitude of circle center, decimal degrees
centerLon = -75.2524474 # Longitude of circle center, decimal degrees

# coordinates = (centerLat,centerLon)
# results = rg.search(coordinates)
# print("reverse geocoder {}".format(results))



# parameters
N = 40 # number of discrete sample points to be generated along the circle

# generate points
circlePoints = []
while radius > jump:
    for k in range(N):
        # compute
        angle = math.pi*2*k/N
        dx = radius*math.cos(angle)
        dy = radius*math.sin(angle)
        point = {}
        point['lat']=centerLat + (180/math.pi)*(dy/6378137)
        point['lon']=centerLon + (180/math.pi)*(dx/6378137)/math.cos(centerLat*math.pi/180)
        # add to list
        circlePoints.append(point)
    radius = radius-jump
    

# print(circlePoints)
```

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[centerLat, centerLon], zoom_start=19, max_zoom=25)

# add markers to map
for cl in circlePoints:
    lat = cl['lat']
    lng  = cl['lon']
    label = '{}, {}'.format(lat, lng)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=2,
        popup=label,
        color='#e60e1c',
        fill=True,
        fill_color='#e60e1c',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
